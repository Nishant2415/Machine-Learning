# Set 3 ( Div - A )

### Load the Diabetes dataset

```python
diabetes = pd.read_csv('https://github.com/Nishant2415/Machine-Learning/raw/master/Dataset/diabetes.csv')
```

```python
diabetes
```

### Delete all duplicate values by preserving the last records

```python
diabetes.drop_duplicates(keep='last')
```

### Build a method where every value should be increased by 10, Apply this function to all the values of Glucose column

```python
def add_glucose(count):
    temp = diabetes['Glucose'] + count
    print(temp)
add_glucose(10)
```

### Find the mean of the ages by appling grouping on class_actual

```python
diabetes.groupby('class_actual')['Age'].mean()
```

### Find the count of the persons whose BMI is grater than 32

```python
diabetes[diabetes['BMI']>32]['BMI'].count()
```

### Calculate the F1 score by calculating the percision and recall for the column class_acual and class_pred

#### ``Precision and Recall``

```python
precision = precision_score(diabetes['class_actual'],diabetes['class_pred'])
recall = recall_score(diabetes['class_actual'],diabetes['class_pred'])
```

#### ``F1 Score``

```python
2 * (precision * recall) / (precision + recall)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelBinarizer, MultiLabelBinarizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
```

```python
heart = pd.read_csv('https://github.com/Nishant2415/Machine-Learning/raw/master/Dataset/HeartDisease.csv')
```

```python
heart.head()
```

```python
heart.drop('class_predict',axis=1,inplace=True)
```

```python
heart.dropna(inplace=True)
```

```python
pca = PCA(n_components = 0.99)
features_pca = pca.fit_transform(heart)
print('Original features',heart.shape[1])
print('Reduced features',features_pca.shape[1])
```

```python
features = heart.iloc[:,:-1]
features.head()
```

```python
lda = LinearDiscriminantAnalysis(n_components=1)
target = heart.iloc[:, -1:]
features_lda = lda.fit(features, np.array(target).ravel()).transform(features)
print('Original features',heart.shape[1])
print('Reduced features',features_lda.shape[1])
```
